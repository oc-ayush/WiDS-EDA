**Import pandas as pd.**

```python
import pandas as pd
```

**Read Salaries.csv as a dataframe called sal.**

```python
sal = pd.read_csv('Salaries.csv', low_memory=False)
```

**Check the head of the DataFrame.**

```python
print(sal.head(5))
```

**Use the .info() method to find out how many entries there are.**

```python
sal.info()
```

**What is the average BasePay ?**

```python
sal['BasePay'] = pd.to_numeric(sal['BasePay'], errors='coerce')
```

```python
average_basepay = sal['BasePay'].mean()
print(f"Average BasePay: {average_basepay}")
```

**What is the highest amount of OvertimePay in the dataset ?**

```python
highest_op = pd.to_numeric(sal['OvertimePay'], errors='coerce').max()

print(highest_op)
```

**What is the job title of  JOSEPH DRISCOLL ? Note: Use all caps, otherwise you may get an answer that doesn't match up (there is also a lowercase Joseph Driscoll).**

```python
sal.loc[sal['EmployeeName'] == 'JOSEPH DRISCOLL', 'JobTitle'].iloc[0]
```

**How much does JOSEPH DRISCOLL make (including benefits)?**

```python
totalpaybenefits = sal.loc[sal['EmployeeName'] == 'JOSEPH DRISCOLL', 'TotalPayBenefits'].iloc[0]
totalpaybenefits = float(totalpaybenefits)
print(totalpaybenefits)
```

**What is the name of highest paid person (including benefits)?**

```python
max_pay = pd.to_numeric(sal['TotalPayBenefits'], errors='coerce').max()

sal.loc[sal['TotalPayBenefits'] == max_pay, 'EmployeeName'].iloc[0]
```

**What is the name of lowest paid person (including benefits)? Do you notice something strange about how much he or she is paid?**

```python
lowest_pay = pd.to_numeric(sal['TotalPayBenefits'], errors='coerce').min()

sal.loc[sal['TotalPayBenefits'] == lowest_pay, 'EmployeeName'].iloc[0]
```

**What was the average (mean) BasePay of all employees per year? (2011-2014) ?**

```python
sal.groupby('Year')['BasePay'].mean()
```

**How many unique job titles are there?**

```python
sal['JobTitle'].nunique()
```

**What are the top 5 most common jobs?**

```python
sal['JobTitle'].value_counts().head(5)
```

**How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?)**

```python
sum(sal[sal['Year'] == 2013]['JobTitle'].value_counts() == 1)
```
